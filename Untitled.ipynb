```python
import pandas as pd
```

```python
penguin = pd.read_csv('penguins_cleaned.csv')
```

```python
data = penguin.copy()
```

```python
target = 'species'
```

```python
encode = ['sex','island']
```

```python
data.tail()
```

```python
for col in encode:
    dummy = pd.get_dummies(data[col], prefix=col)
    data = pd.concat([data,dummy], axis=1)
    del data[col]
```

```python
data
```

```python
target_mapper = {'Adelie':0, 'Chinstrap':1, 'Gentoo':2}
```

```python
def target_encode(val):
    return target_mapper[val]
```

```python
data['species'] = data['species'].apply(target_encode)
```

```python
data
```

```python
X = data.drop('species', axis=1)
Y = data['species']
```
